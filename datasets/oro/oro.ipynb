Acerca del conjunto de datos
Contexto

Históricamente, el oro se ha utilizado como forma de moneda en varias partes del mundo, incluidos los EE. UU. En la actualidad, los metales preciosos como el oro se mantienen en los bancos centrales de todos los países para garantizar el pago de las deudas externas y también para controlar la inflación que refleja la solidez financiera del país. Recientemente, las economías emergentes del mundo, como China, Rusia e India, han sido grandes compradoras de oro, mientras que EE. UU., SoUSA, Sudáfrica y Australia se encuentran entre los grandes vendedores de oro.

Pronosticar subidas y bajadas en las tasas diarias del oro puede ayudar a los inversores a decidir cuándo comprar (o vender) la materia prima. Pero los precios del oro dependen de muchos factores, como los precios de otros metales preciosos, los precios del petróleo crudo, el rendimiento de la bolsa de valores, los precios de los bonos, los tipos de cambio de divisas, etc.

El desafío de este proyecto es predecir con precisión el precio de cierre ajustado futuro del ETF de oro durante un período de tiempo determinado en el futuro. El problema es un problema de regresión, porque el valor de salida, que es el precio de cierre ajustado en este proyecto, es un valor continuo.
Contenido

Los datos para este estudio se recopilaron del 18 de noviembre de 2011 al 1 de enero de 2019 de varias fuentes. Los datos tienen 1718 filas en total y 80 columnas en total. Datos de atributos, como el precio del petróleo, el índice Standard and Poor's (S&P) 500, las tasas de los bonos estadounidenses del índice Dow Jones (10 años), los tipos de cambio del dólar estadounidense, los precios de los metales preciosos plata y platino y otros metales como el paladio y el rodio, Se recopilaron los precios del índice del dólar estadounidense, Eldorado Gold Corporation y Gold Miners ETF.

Los datos históricos de Gold ETF obtenidos de Yahoo Finance tienen 7 columnas, Fecha, Apertura, Máximo, Mínimo, Cierre, Cierre ajustado y Volumen, la diferencia entre Cierre ajustado y Cierre es que el precio de cierre de una acción es el precio de ese acción al cierre del día de negociación. Mientras que el precio de cierre ajustado tiene en cuenta factores como dividendos, divisiones de acciones y nuevas ofertas de acciones para determinar un valor. Por lo tanto, el cierre ajustado es la variable de resultado, que es el valor que debe predecir.
