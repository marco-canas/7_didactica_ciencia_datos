Dado que el conjunto de datos no es demasiado grande, puede calcular fácilmente el *coeficiente de correlación estándar* (también llamado r de Pearson) entre cada par de atributos mediante el método `corr()`:

```python
url = 'https://raw.githubusercontent.com/marco-canas/taca/main/datasets/housing/housing.csv'
import pandas as pd 
housing = pd.read_csv(url) 
```

```python
housing.head() 
```

```python
corr_matrix = housing.corr() 
```

```python
corr_matrix 
```

Ahora veamos cuánto se correlaciona cada atributo con `median_house_value` (valor medio de la vivienda): 

```python
corr_matrix["median_house_value"].sort_values(ascending=False)
```

El coeficiente de correlación varía de –1 a 1.

* Cuando está cerca de 1, significa que hay una fuerte correlación lineal positiva;   
* por ejemplo, el valor medio de la vivienda tiende a aumentar cuando aumenta el ingreso medio.

Cuando el coeficiente se acerca a –1, significa que existe una fuerte correlación lineal negativa;   
* puede ver una pequeña correlación negativa entre la latitud y el valor medio de la vivienda (es decir, los precios tienen una ligera tendencia a bajar cuando va al norte).

Finalmente, coeficientes cercanos a 0 significan que no existe correlación lineal.

La Figura 2-14 muestra varias gráficas junto con el coeficiente de correlación entre sus ejes horizontal y vertical.

### ADVERTENCIA

* El coeficiente de correlación solo mide las correlaciones lineales   
* (“si x sube, entonces, $y$ generalmente sube/baja”).

* Observe cómo todas las gráficas de la fila inferior tienen un coeficiente de correlación igual a 0,  
* a pesar de que las variables claramente no son independientes: estos son ejemplos de relaciones no lineales.

* Además, la segunda fila muestra ejemplos donde el coeficiente de correlación es igual a 1 o –1;  
* observe que esto no tiene nada que ver con el valor absoluto de la pendiente.

Por ejemplo, su altura en pulgadas tiene un coeficiente de correlación de 1 con su altura en pies o en
nanómetros.

* Otra forma de verificar la correlación entre atributos es usar la función pandas `scatter_matrix()`, 
* que traza cada atributo numérico contra todos los demás atributos numéricos.

 
* Como tenemos 10 atributos numéricos, obtendría $10^{2} = 100$ gráficos, que no cabrían en   
  una página,  
* así que simplemente nos centraremos en algunos atributos prometedores que parecen estar más   
  correlacionados con el valor medio de la vivienda:

```python
import matplotlib.pyplot as plt 
from pandas.plotting import scatter_matrix

attributes = ["median_house_value", "median_income", "total_rooms", "housing_median_age"] 
scatter_matrix(housing[attributes], figsize=(12, 8))

plt.show() 
```

```python
housing[attributes].corr() 
```

La diagonal principal (de arriba a la izquierda a abajo a la derecha) estaría llena de líneas rectas si Pandas trazara cada variable contra sí misma, lo que no sería muy útil.

* Entonces, en su lugar, Pandas muestra un histograma de cada atributo   
* (hay otras opciones disponibles; consulte la documentación de pandas para obtener más detalles).

El atributo más prometedor para predecir el valor medio de la vivienda es el ingreso medio, así que ampliemos su diagrama de dispersión de correlación:

```python
housing.plot(kind="scatter", x="median_income", y="median_house_value",
alpha=0.1, figsize = (10,5))

plt.savefig('median_house_value_vs_median_income.jpg') 
```

Esta trama revela algunas cosas.   
1. la correlación es realmente muy fuerte; se puede ver claramente la tendencia alcista y   
   los puntos no están demasiado dispersos.

2. En segundo lugar, el límite de precio que notamos anteriormente es claramente visible como   
   una línea horizontal en $ 500.000.

Pero este diagrama de dispersión revela otras líneas rectas menos obvias: una línea horizontal alrededor de $\$450,000$, otra alrededor de $\$350,000$, quizás una alrededor de $\$ 280,000$ y algunas más por debajo de eso.
