```python
log_reg=LogisticRegression()  
```

```python
import matplotlib.pyplot as plt 
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1) 
```

```python
y_proba = log_reg.predict_proba(X_new) 
```

```python
plt.plot(X_new,y_proba[:,1],'g-',label='Iris Virginica')
plt.plot(X_new,y_proba[:,0],'b--',label='Not Iris Virginica')
plt.legend() 
plt.grid() 
```

```python
from sklearn import datasets 
```

```python
iris=datasets.load_iris() 
```

```python
iris.keys()  
```

```python
iris['feature_names'] 
```

```python
X=iris['data'][:,3:] 
```

```python
iris['target_names'] 
```

```python
y = (iris['target']==2).astype(np.int)  
```

```python
y 
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
log_reg = LogisticRegression()  
```

```python
log_reg.fit(X,y)   
```

```python
(min(X),max(X)) 
```

```python
X_new=np.arange(0.1,2.5,0.01).reshape(-1,1)  
```

```python
y_new_posi = log_reg.predict_proba(X_new)[:,1] 
```

```python
y_new_neg = log_reg.predict_proba(X_new)[:,0] 
```

```python
y_new_neg 
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(X_new.ravel(),y_new_neg,X_new,y_new_posi)   
```
