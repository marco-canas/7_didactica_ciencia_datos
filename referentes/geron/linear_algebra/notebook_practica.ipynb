```python
r = np.array([9,-1])
s = np.array([7,-3]) 
```

# Vectores 2D como flechas

```python
def trazar_vec_2d(vector_2d, inicio_vec = [0,0],  **kwords):
    return plt.arrow(inicio_vec[0], inicio_vec[1], vector_2d[0], vector_2d[1], 
                    head_width = 0.2, head_length = 0.3, **kwords) 
```

```python
trazar_vec_2d(np.array([9,-1]), color = 'r') 
trazar_vec_2d(s, color = 'b') 
plt.axis([-1,11,-5,1])
plt.grid(True, alpha = 0.3) 
plt.show() 
```

## Práctica de trazado de vectores 3D

## Objetivo: adquirir la metodología Geron para el trazado de vectores 3D

```python
import numpy as np 
%matplotlib notebook 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
r = np.array([1,2,3])
s = np.array([4,5,6]) 
```

```python
def trazado_vectores_3d(ax, vectores, **kwords):
    x_cor, y_cor, z_cor = zip(*vectores)
    ax.scatter(x_cor,y_cor, z_cor, **kwords)
    for v in vectores:
        x,y,z = v
        ax.plot([x,x], [y,y], [0,z], color = 'gray', linestyle = 'dotted', marker = '.')
```

```python
subplot3d = plt.subplot(111, projection = '3d')
trazado_vectores_3d(subplot3d, [r,s], color = ['r', 'b']) 
subplot3d.set_zlim([0,8])
plt.show() 
```
