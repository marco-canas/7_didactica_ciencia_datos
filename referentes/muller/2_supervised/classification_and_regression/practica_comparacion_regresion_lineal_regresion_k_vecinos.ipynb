# Práctica de comparación gráfica entre la regresión lineal y la regresión con el algoritmo de los K - vecimos más cercanos 

```python
import numpy as np 
```

```python
X = np.arange(0,1,0.02)      
X 
```

```python
y = 2*X + 1 + 0.5*np.random.randn(50)
```

```python
y 
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X,y, color = 'yellow')  
plt.grid(True) 
plt.style.use('dark_background') 
plt.title('Conjunto de datos artificial') 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42, train_size = 0.80 ) 
```

```python
X_train = X_train.reshape(-1,1) 
X_test = X_test.reshape(-1,1) 
```

```python
from sklearn.linear_model import LinearRegression
reg_lin = LinearRegression() 
```

```python
reg_lin.fit(X_train,y_train) 
```

```python
reg_lin.coef_, reg_lin.intercept_
```

```python
fig, ax = plt.subplots(figsize = (8,8)) 

ax.plot(X,reg_lin.predict(X.reshape(-1,1)))  
ax.plot(X_train, y_train, '^', color = 'yellow') 
ax.plot(X_test, y_test, 'v', color = 'green') 
```

```python
reg_lin.score(X_train, y_train), reg_lin.score(X_test,y_test) 
```

```python
from sklearn.neighbors import KNeighborsRegressor
reg_knn = KNeighborsRegressor(n_neighbors= 6) 
```

```python
reg_knn.fit(X_train,y_train) 
```

```python
fig, axes = plt.subplots(1,2,figsize = (20,10))    

axes[0].plot(X,reg_lin.predict(X.reshape(-1,1)))  
axes[0].plot(X_train, y_train, '^', color = 'yellow') 
axes[0].plot(X_test, y_test, 'v', color = 'green') 

axes[1].plot(X,reg_knn.predict(X.reshape(-1,1)))  
axes[1].plot(X_train, y_train, '^', color = 'yellow') 
axes[1].plot(X_test, y_test, 'v', color = 'green') 

plt.savefig('comparacion.jpg')
```

```python
reg_knn.score(X_train,y_train), reg_knn.score(X_test,y_test) 
```
