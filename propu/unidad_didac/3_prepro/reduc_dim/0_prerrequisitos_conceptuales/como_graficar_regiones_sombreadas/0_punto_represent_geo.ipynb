## Un punto en en un espacio de una dimensión

### Secuencia de esta actividad

1. importe las librerías numpy y matplotlib 
2. crear una figura y adicionarle dos ejes coordenados 
3. Trace una representación del espacio 1 - dimensional
4. Trace el intervalo $[0,1]$ de este espacio.
5. Trace un punto de este intervalo.

```python
# 1. importe las librerías numpy y matplotlib 
import numpy as np 
import matplotlib.pyplot as plt 

plt.style.use('dark_background') 
# 2. crear una figura y adicionarle dos ejes coordenados 
fig = plt.figure(figsize=(8,4)) 
ax = fig.add_subplot()

# 3. Trace una representación del espacio 1 - dimensional
ax.plot([-3,3],[0,0], color = 'blue', alpha = 0.4)   

# 4. Trace el intervalo $[0,1]$ de este espacio.
ax.plot([0,1],[0,0], color = 'yellow', alpha = 0.8)   

# 5. Trace un punto de este intervalo.
p = np.random.rand(1,1)  
ax.scatter(p,[0], color = 'red', linewidth = 8.0, alpha = 1.0) 
 
```

## Un punto en un espacio de dos dimensiones

```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.style.use('dark_background') 

fig = plt.figure()
ax = fig.add_subplot()

x1 = np.linspace(0,1,10)
y1 = np.zeros_like(x1)

x2 = x1
y2 = np.ones_like(x1) 

x = np.vstack((x1,x2))
y = np.vstack((y1,y2)) 

ax.plot(x,y, color = 'blue', alpha = 0.6) 

plt.axis('equal') 

np.random.seed(513) 

p = np.random.rand(1,2)
ax.scatter(p[:,0],p[:,1], color = 'red', linewidth = 8.0, alpha = 1.0)   

print(p) 
plt.show() 
```

## Punto aleatorio del intervalo $[0,1]$ en el espacio euclidiano 3 - dimensional 

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import axes3d  
%matplotlib notebook  
plt.style.use('dark_background') 

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

T = np.linspace(0,1,2)
x,y = T,T
X,Y = np.meshgrid(x,y) 

Z = np.ones_like(X)

for t in T:
    ax.plot_surface(X,Y,Z*t, color = 'blue', alpha = 0.6)  

    
np.random.seed(513) 
p = np.random.rand(1,3)
ax.scatter(p[:,0], p[:,1], p[:,2], color = 'red', linewidth = 6.0, alpha = 1.0)   
plt.savefig('punto_en_tres_dim.png')


```

# Algoritmo para crear la función distancia promedio entre dos puntos como función de la dimensión  

```python
dim = [1,2,3,10,100,200,500,800,1000] 
```

```python
len(dim) 
```

```python
dist_prom = [] 
for d in dim: 
    p = np.random.rand(10000,d) 
    # esta es una matriz en la que cada fila es un punto del intervalo [0,1]
    distancias = []
    for i in range(10000-1):
        distancias.append(np.linalg.norm(p[i,:] - p[i+1,:]))
    dist_prom.append(np.mean(distancias))     

np.around(dist_prom, decimals=2)       
```

```python
plt.plot(dim, dist_prom) 
```
