```python
import numpy as np
import matplotlib.pyplot as plt  
 
fig = plt.figure()
 
ax = fig.add_subplot()
 
# parametrización de la circunferencia unitaria 
theta = np.linspace(0, 2*np.pi,800)  
x = np.cos(theta)
y = np.sin(theta) 

# Trazar el anillo 
v = np.linspace(0.8,1,2) 
v.shape = (2, 1)
x1 = v * x
y1 = v * y
ax.plot(x1, y1, color='black', linewidth = 2)  
plt.show() 
```

```python
# segundo anillo 
x1 = v * x + 2.2
y1 = v * y
ax.plot(x1, y1, color='red', linewidth = 2)  

# tercer anillo 
x1 = v * x - 2.2
y1 = v * y
ax.plot(x1, y1, color='blue', linewidth = 2)   


# cuarto anillo 
x1 = v * x  + 1.1
y1 = v * y - 1.0
ax.plot(x1, y1, color='green', linewidth = 2)  

# quinto anillo 
x1 = v * x  - 1.1
y1 = v * y - 1.0
ax.plot(x1, y1, color='yellow', linewidth = 2)  



plt.axis('equal')  
plt.show()
```
