## 1 Diagramas de dispersión  

Los gráficos de dispersión muestran cómo las características se relacionan entre sí, el tema principal entre todos los tipos de gráficos relacionales es que muestran cómo las entidades están interconectadas entre sí. Esto se puede hacer en muchos tipos diferentes de parcelas.


```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns 
import warnings 
warnings.simplefilter('ignore')
```

```python
url = 'https://raw.githubusercontent.com/marco-canas/taca/main/datasets/dm_office_sales.csv'
df = pd.read_csv(url) 
df.head() 
# este dataset se encuentra en Kaggle a través del vínculo: https://www.kaggle.com/notepub/dm-office-sales
```

```python
sns.scatterplot(x = 'salary', y = 'sales', data = df) 
plt.show() 
```

### Connecting to Figure in Matplotlib

```python
plt.figure(figsize = (8,4)) 
sns.scatterplot(x = 'salary', y = 'sales', data = df) 
plt.show() 
```

### hue  
The hue and palette parameters are commonly available around many plot calls in Seaborn.  
Color points based off a categorical feature in the DataFrame. 

```python
plt.figure(figsize = (8,4)) 
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'division') 
plt.show() 
```
