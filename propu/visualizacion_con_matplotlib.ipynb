# `.plot(x, y)`

```python
#@title Sobre `.plot()` como "linea poligonal" con ilusión óptica de curva { run: "auto", vertical-output: true }
n = 18 #@param {type:"slider", min:1, max:20, step:1}

import matplotlib.pyplot as plt
import numpy as np
#
x = np.linspace(-2, 2, n)
#
y = x**2
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
#@title Sobre `xlim`, `ylim`, `xticks` y `yticks`: parámetros del método .set() para delimitar valores de los ejes y las marcas sobre los ejes { run: "auto", vertical-output: true }
x_max = 8 #@param {type:"slider", min:0, max:10, step:1}
y_max = 10 #@param {type:"slider", min:0, max:10, step:1}
x_max_ticks = 6 #@param {type:"slider", min:0, max:10, step:1}
y_max_ticks = 7 #@param {type:"slider", min:0, max:10, step:1}


import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()
ax.plot(x, y)
#
ax.set(xlim=(0, x_max),   
       xticks=np.arange(1, x_max_ticks),  
       ylim=(0, y_max),  
       yticks=np.arange(1, y_max_ticks))
#
plt.show()
```

```python
#@title Sobre `.plot(..., linewidth=float)` para controlar grosor de línea. Sobre `.grid()` método que espera True o False para controlar salida de cuadrícula { run: "auto", vertical-output: true }
linewidth = 15 #@param {type:"slider", min:0, max:15, step:1}
booleano_cuadricula =  True#@param {type:"raw"}

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)
######### 
fig, ax = plt.subplots()
ax.grid(booleano_cuadricula)
ax.plot(x, y, linewidth=linewidth)
#########
plt.show()
```

```python
#@title Sobre `ax.title()`, `ax.xlabel()` y `ax.ylabel()` que reciben strings y controlan salida de título y etiquetas para los ejes { run: "auto", vertical-output: true }
title = " Mi t\xEDtulo" #@param {type:"string"}
xlabel = "nuevo eje horizontal" #@param {type:"string"}
ylabel = " variable y" #@param {type:"string"}


import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()
ax.set_title(title)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.plot(x, y, linewidth=2)
plt.show()
```

## `.plot(..., marker=str)` 

```python
#@title `.plot(..., marker=str)` para controlar estilos de puntos tales como ".", "o", "v", "^", "<", "8", "s", "P", "*", "H", "+", "X", "D" { run: "auto", vertical-output: true }
estilo = "D" #@param {type:"string"}
p1 = [3, 4] #@param {type:"raw"}
p2 = [0, 0] #@param {type:"raw"}
p3 = [4, -1] #@param {type:"raw"}
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
puntos = np.array([p1, p2, p3, p1]).reshape(4,2)
ax.plot(puntos[:, 0], puntos[:, 1], marker=estilo)
ax.grid(True)
plt.show()
```

```python
#@title `.plot(x, y1, str, x, y2, str, x, y3, str ...)` para controlar estilos indicando color con una letra y tipo de marker { run: "auto", vertical-output: true }
estilos = ["r--", "b-", "g."] #@param {type:"raw"}
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2)
y1, y2, y3 = x, x**2, x**3

fig, ax = plt.subplots()
ax.plot(x, y1, estilos[0], x, y2, estilos[1], x, y3, estilos[2])
plt.show()
```

# `.scatter()`

```python
#@title sobre `.scatter(..., s=iterable_de_sizes, c=iterable_de_colores)` para controlar tamaños y colores { run: "auto", vertical-output: true }
s = 25 #@param {type:"slider", min:1, max:30, step:4.0}

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

sizes = s * np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
#@title `.scatter(..., alpha=float)` para controlar opacidad { run: "auto", vertical-output: true }
alpha = 0.1 #@param {type:"slider", min:0, max:1, step:0.1}
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y = 0.5*x + 5 + np.random.randn(1000)*2

fig, ax = plt.subplots()
ax.set(xlim=(-2.5, 2.5), ylim=(-5, 15))
plt.scatter(x, y, alpha=alpha)
plt.show()
```

```python
#@title `.plot()` que traza línea de tendencia { run: "auto", vertical-output: true }
pendiente = 1.6 #@param {type:"slider", min:-2, max:2, step:0.2}
intercepto = -1 #@param {type:"slider", min:-2, max:2, step:1}

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y = pendiente * x + intercepto + np.random.randn(1000)*2

fig, ax = plt.subplots()

ax.set(xlim=(-2.5, 2.5), ylim=(-10, 10))
ax.scatter(x, y, alpha=0.3)   
xmin=-2.5
xmax=2.5
ax.plot([xmin, xmax],
         [xmin*pendiente+intercepto, xmax*pendiente+intercepto],
         linewidth=4, color="red")
ax.grid(True)

plt.show()
```

# `.meshgrid()`

```python
#@title `.meshgrid()` para crear regilla de puntos { run: "auto", vertical-output: true }
n = 5 #@param {type:"slider", min:2, max:5, step:1}
m = 5 #@param {type:"slider", min:2, max:5, step:1}
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, n, n)
y = np.linspace(1, m, m)
x_matriz, y_matriz = np.meshgrid(x, y)
pares_ordenados = np.hstack([x_matriz.ravel().reshape(-1,1), y_matriz.ravel().reshape(-1, 1)])

def grafica_ejes(a, b):
    axe = plt.gca()
    axe.plot([-0.5, a], [0,0])
    axe.plot([0,0], [-0.5, b])
    axe.set(xlim=(min(x)-2, max(x)+2), ylim=(min(y)-2, max(y)+2))

fig, ax = plt.subplots(1, 3, figsize=(15,5))
plt.subplot(131)
plt.plot(x, np.zeros_like(x), "o")
for par in pares_ordenados:
  a, b = par[0], par[1]
  plt.annotate(f"{int(a)}", xy=(a,b), xytext=(a, b-0.3),
              arrowprops=dict(color='white',  shrink=0.5))
plt.plot(pares_ordenados[:, 0], pares_ordenados[:, 1], "D")
plt.legend(["x 1D", "x_matrix 2D"])
plt.title("x 1D y x_matrix 2D")
grafica_ejes(max(x), max(y))
plt.subplot(132)
plt.plot(np.zeros_like(y), y, "o")
for par in pares_ordenados:
  a, b = par[0], par[1]
  plt.annotate(f"{int(b)}", xy=(a,b), xytext=(a, b-0.3),
              arrowprops=dict(color='white',  shrink=0.5))
plt.plot(pares_ordenados[:, 0], pares_ordenados[:, 1], "D")
plt.legend(["y 1D", "y_matrix 2D"])
plt.title("y 1D y y_matriz 2D")
grafica_ejes(max(x), max(y))
plt.subplot(133)
for par in pares_ordenados:
  a, b = par[0], par[1]
  plt.annotate(f"({int(a)},{int(b)})", xy=(a,b), xytext=(a-0.3, b-0.3),
              arrowprops=dict(color='white',  shrink=0.5))
plt.plot(pares_ordenados[:, 0], pares_ordenados[:, 1], "D")
plt.title("Representación de np.meshgrid(x, y)")
grafica_ejes(max(x), max(y))
plt.show()

print("x 1D es: \n", x)
print("x_matriz 2D es: \n", x_matriz)
print("y 1D es: \n", y)
print("y_matriz 2D es: \n", y_matriz)
```

# `.imshow()`

```python
#@title `.meshgrid()` para crear regilla de puntos { run: "auto", vertical-output: true }
n = 18 #@param {type:"slider", min:2, max:50, step:1}
m = 29 #@param {type:"slider", min:2, max:50, step:1}
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, n, n)
y = np.linspace(1, m, m)
x_matriz, y_matriz = np.meshgrid(x, y)
pares_ordenados = np.hstack([x_matriz.ravel().reshape(-1,1), y_matriz.ravel().reshape(-1, 1)])

def grafica_ejes(a, b):
    axe = plt.gca()
    axe.plot([-0.5, a], [0,0])
    axe.plot([0,0], [-0.5, b])
    axe.set(xlim=(min(x)-2, max(x)+2), ylim=(min(y)-2, max(y)+2))

fig, ax = plt.subplots(1, 3, figsize=(15,5))
plt.subplot(131)
plt.plot(pares_ordenados[:, 0], pares_ordenados[:, 1], "D")
plt.plot(x, np.zeros_like(x), "o")
plt.legend(["x 1D", "x_matrix 2D"])
plt.title("x 1D y x_matrix 2D")
grafica_ejes(max(x), max(y))
plt.subplot(132)
plt.plot(np.zeros_like(y), y, "o")

plt.plot(pares_ordenados[:, 0], pares_ordenados[:, 1], "D")
plt.legend(["y 1D", "y_matrix 2D"])
plt.title("y 1D y y_matriz 2D")
grafica_ejes(max(x), max(y))
plt.subplot(133)

# plt.plot(pares_ordenados[:, 0], pares_ordenados[:, 1], "s")
plt.title("color dependiendo de posición")
Z = x_matriz ** 2 + y_matriz ** 2
plt.imshow(Z)
grafica_ejes(max(x), max(y))
plt.show()
```

# `.contour()`
